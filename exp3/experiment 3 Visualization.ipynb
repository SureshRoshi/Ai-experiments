```python
# import the packages

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# import the data
df=pd.read_csv("students_placement_data.csv")
df.head()
```

### 1) Line Graphs 
Draw a line graph between SSC and intermediate/diploma percentage. Does it make sense? Explain when to use line graphs?

```python
plt.plot(df["SSC Percentage"],df["inter_Diploma_percentage"])
```

I think line graph should not be used here.A line chart or line plot or line graph or curve chart is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.

## 2) Scatter Plot 
Draw a scatter plot between SSC and inter/diploma percentage. Do you find any correlation by just visual inspection? If yes, Then what kind of correlation you can see? (+ve, -ve or No)

```python
x=df["SSC Percentage"]
y=df["inter_Diploma_percentage"]
plt.scatter(x,y)
plt.xlabel("SSC scores")
plt.ylabel("inter/diploma scores")
plt.title("Scatter plot between SSC and inter/diploma scores")
```

#### It is clear thar these two variables are possitively correlated

## 3) Histogram
Draw a histogram for B.Tech percentage. How many people have scores between 60-80% in B.tech. Tell your approximate answer by visual inspection.

```python
bins=np.linspace(0,100,11) # Generates 6 numbers between 0 and 10 which are equally spaced. 
print(bins)

plt.hist(df["B.Tech_percentage"],bins,width=9)
```

#### Approximatly 65 people are having score between 60-80% in B.tech.

## 4) Subplots

Draw a scatter plot between 
* Intermediate and SSC 
* Intermediate and B.Tech
in a same graph.

What's your observation?

```python
fig,ax=plt.subplots()
ax.scatter(df["inter_Diploma_percentage"],df["SSC Percentage"], label="Inter-SSC")
ax.scatter(df["inter_Diploma_percentage"],df["B.Tech_percentage"],label="Inter-B.Tech")
ax.set_xlabel("Intermediate Score") 
ax.legend(loc=2) 
```

### People who tend to get good marks in intermediate also got good percentage in SSC. But, Their marks have been reduced in B.tech as compared to SSC.

## 5) jointplot
Draw a jointplot between intermediate and Backlogs in B.tech. What's your observation.

```python
sns.jointplot(df["inter_Diploma_percentage"],df["Backlogs"],data=df) 
```

#### 1) There are more people with zero backlogs
#### 2) People who got good percentage in intermediate have less backlogs (Comparatively) .
#### 3) Majority of people got 75+ percentage in intermediate

Draw a heat map between SSC, Intermediate and B.Tech percentage and find the correlation between them.
